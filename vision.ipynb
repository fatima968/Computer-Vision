```python
from deepface import DeepFace
import cv2

# Captura desde la cámara
cap = cv2.VideoCapture(0)
ret, frame = cap.read()
cv2.imwrite("captura.jpg", frame)
cap.release()

# Analiza la imagen
analysis = DeepFace.analyze(img_path="captura.jpg", actions=['age','gender','emotion'])
print(analysis)
```
